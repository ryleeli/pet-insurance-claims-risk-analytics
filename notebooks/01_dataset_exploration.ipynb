# Check the distribution of compensation amounts

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df["loss"], bins=100)
plt.title("Claim Severity Distribution")
plt.xlabel("Loss")
plt.ylabel("Frequency")

plt.tight_layout()

plt.savefig(
    "../images/claim_severity_distribution.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
